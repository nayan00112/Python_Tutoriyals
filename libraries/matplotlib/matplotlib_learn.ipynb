```python
x = ["apple", "mango", "pineapple", "banana", "orange", "guava"]
y = [80, 60, 100, 50, 70, 65]

color=["red", "yellow", "#555500","yellow", "orange", "green"]
plt.bar(x, y, color=color)
```

```python
label = ["apple", "mango", "pineapple", "banana", "orange", "guava"]
y = [80, 60, 100, 50, 70, 65]

plt.pie(y, labels=label)
plt.show()
```

```python
color= ["#"+str(np.random.randint(9))+str(np.random.randint(9))+str(np.random.randint(9)) for i in label]
print(color)
plt.pie(y, labels=label, colors=color)
plt.show()
```
